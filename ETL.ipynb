<h2> Vamos a empezar revisar los df y hacer cambios necesarios para mejor trabajo </h2>

```python
# Revisión de la información de los df
df_pen_hogares.info()
df_pen_poblacion.info()
```

```python
# Combinamos los df considerando las columnnas "Año", "Trimestre" y "Provincia"
df_combinado_penetracion = pd.merge(df_pen_poblacion, df_pen_hogares, on=['Año', 'Trimestre', 'Provincia'])
df_combinado_penetracion.info()
df_combinado_penetracion.head()
```

```python
# Revisión de la información del df Acceso de tecnologia por provincia:
df_acc_tec_prov.info()
df_acc_tec_prov.head()
```

```python
# Al revisar este df vemos que hay datos con errores
df3 = df_acc_tec_prov.copy()
```

```python
# Vamos a sacar los "*" que contienen valores en la columna Año y Trimestre
df3['Año'] = df3['Año'].apply(lambda x: x.replace('*', '') if isinstance(x, str) else x)
df3['Trimestre'] = df3['Trimestre'].apply(lambda x: x.replace('*', '') if isinstance(x, str) else x)
```

```python
# Filtra las filas en 'Trimestre' que no son cadenas de texto y convertimos a INT la columna
df3 = df3[pd.to_numeric(df3['Trimestre'], errors='coerce').notna()]
df3['Trimestre'] = df3['Trimestre'].astype(int)

# Pasamos la columna año a INT
df3['Año'] = df3['Año'].astype(int)
```

```python
df3.info()
```

```python
# Combinamos el df corregido con el previamente obtenido, considerando las mismas columnas
df_combinado2 = pd.merge(df3, df_combinado_penetracion, on=['Año', 'Trimestre', 'Provincia'])
df_combinado2.info()
```

<h2> Transformación del primer df: </h2>

```python
df_combinado.sample(5)
```

```python
df_combinado.info()
```

```python
# Comparar los valores de las columnas y contar los iguales
cantidad_iguales1 = (df_combinado['Total'] == df_combinado['Total_x']).sum()
cantidad_iguales2 = (df_combinado['Total'] == df_combinado['Total_y']).sum()
cantidad_iguales3 = (df_combinado['Total_x'] == df_combinado['Total_y']).sum()

# Imprimir la cantidad de valores iguales
print(cantidad_iguales1)
print(cantidad_iguales2)
print(cantidad_iguales3) # en este caso coinciden todos los valores
```

```python
# Rellenar los valores NaN con 0
df_combinado = df_combinado.fillna(0)

# Convertir las columnas de float a int
df_combinado['Accesos por cada 100 hab'] = df_combinado['Accesos por cada 100 hab'].astype(int)
df_combinado['Accesos por cada 100 hogares'] = df_combinado['Accesos por cada 100 hogares'].astype(int)
```

#### Podemos notar que hay diferencia entre los resultados "Totales" procedentes de los distintos dataset originales.
#### En este caso, "Total_x" = "Total_y" (que suman los diferentes tipos de conexiones) pero difieren respecto a "Total" (procedente de la suma de Mbps) Como coinciden "Total_x" con "Total_y", se elimina la ultima columna.

```python
# Eliminamos la ultima columna
df_combinado = df_combinado.drop('Total_y', axis=1)
df_combinado.head()
```

```python
# Renombramos las columnas necesarias
df_combinado = df_combinado.rename(columns={'Total': 'Total suma Mbps'})
df_combinado = df_combinado.rename(columns={'Total_x': 'Total_conexion'})
df_combinado = df_combinado.rename(columns={'OTROS': 'Otros Mbps'})
df_combinado = df_combinado.rename(columns={'HASTA 512 kbps': 'Hasta 512 kbps'})
```

```python
# Revisamos el nuevo df obtenido
df_combinado.head()
```

```python
# Agrupar y resumir los datos:
df_agrupado = df_combinado.groupby(['Año', 'Trimestre', 'Provincia']).sum()
df_agrupado.sample(10)
```

<h2> Transformación del Segundo df: </h2>

```python
df_total_nacional.info()
```

```python
# Se eliminan las columnas
df_total_nacional = df_total_nacional.drop(['Periodo_x', 'Trimestre.1', 'Periodo_y', 'Periodo'], axis=1)
df_total_nacional.info()
```

```python
df_total_nacional.head(10)
```

```python
# Se revisan los valores que coindiden:
cantidad_iguales = (df_total_nacional['Total_x'] == df_total_nacional['Total_y']).sum()
# Imprimir la cantidad de valores iguales
print(cantidad_iguales)
```

```python
# Existen valores con "." en algunas columnas, se procede a retirarlos
df_total_nacional['Hasta 512 kbps'] = df_total_nacional['Hasta 512 kbps'].astype(str).str.replace('.', '')
df_total_nacional.head(5)
```

```python
# Eliminamos del mismo modo "." del resto de columnas
columnas = ['Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps', 'OTROS']

for columna in columnas:
    df_total_nacional[columna] = df_total_nacional[columna].astype(str).str.replace('.', '')
```

```python
# Revisamos las similitudes de las columnas
cantidad_iguales5 = (df_total_nacional['Total'] == df_total_nacional['Total_x']).sum()
# Imprimir la cantidad de valores iguales
print(cantidad_iguales5)
```

```python
# Eliminamos columnas con información redundante
df_total_nacional = df_total_nacional.drop(['Total_y', 'Total_x'], axis=1)
df_total_nacional.head(3)
```

```python
# Renombramos las columnas
df_total_nacional = df_total_nacional.rename(columns={'OTROS': 'Otros Mbps'})
df_total_nacional = df_total_nacional.rename(columns={'Otros': 'Otras conexiones'})
df_total_nacional = df_total_nacional.rename(columns={'Total': 'Total Mbps-conexiones'})
```

```python
df_total_nacional.info()
```

```python
# Se convierten columnas en float
columnas_convertir = ['Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps', 'Otros Mbps']
df_total_nacional[columnas_convertir] = df_total_nacional[columnas_convertir].astype(float)
```

```python
df_total_nacional.info()
```
